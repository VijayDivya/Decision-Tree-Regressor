```python
import pandas as pd
```

```python
#loading dataset
dataset=pd.read_csv("50_Startups.csv")
```

```python
dataset
```

```python
#dummies used to avoid repetition of information and save memory
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
# to display name of the columns in dataset
dataset.columns
```

```python
# assigning input variables
independent=dataset[['R&D Spend', 'Administration', 'Marketing Spend', 'State_Florida', 'State_New York']]
```

```python
independent
```

```python
# Assigning output variables
dependent=dataset['Profit']
```

```python
dependent
```

```python
# Spliting training set & test set
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
# model creation
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor(criterion='squared_error', splitter='best', max_depth='int')
regressor=regressor.fit(X_train,Y_train)
```

```python
# To create graph-like tree
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
# Evaluation using test set
Y_pred=regressor.predict(X_test)
```

```python
# Finding r^2 value/score
from sklearn.metrics import r2_score
r_score=r2_score(Y_test,Y_pred)
```

```python
r_score
```

```python
# To save model 
import pickle

# To create filename
filename="finalized_model_dt_linear.sav"

#  Model saved to the filename
pickle.dump(regressor,open(filename,'wb'))
```

```python
# To load/read the saved model and check result/output
loaded_model=pickle.load(open("finalized_model_dt_linear.sav",'rb'))
result=loaded_model.predict([[1234,345,4565,1,0]])
```
